# 군집분석을 위해 데이터셋 가공

```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/응통 분석 공모전/데이터/"
```

```python
data = pd.read_csv(path + "data.csv")
data
```

```python
data_cluster = data.drop(['자치구','행정동', 'DONG_recycle'], axis=1)
data_cluster
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(data_cluster)
data_cluster_transformed=scaler.transform(data_cluster)
```

```python
X = data_cluster_transformed
X
```

# K-means

* n_clusters : 사전으로 정의하는 군집 갯수로서 K-means가 몇 개의 군집으로 군집화할 것인지 즉, 'K'값을 정의해주는 것이다.
* init : 초기에 군집 중심점의 좌표를 설정하는 방식으로 보통 'k-means++' 방식으로 설정.
* max_iter : 반복횟수를 설정하는 것이다. 설정한 횟수에 도달하기 이전에 군집화가 완료되면 반복횟수를 채우지 않더라도 중간에 종료된다.

```python
from sklearn.cluster import KMeans

# 1.K-means 할당
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=1000)

labels = kmeans.fit_predict(X)

data['kmeans_cluster'] = labels

data.head()
```

```python
# 전체 실루엣 계수 계산
from sklearn.metrics import silhouette_score, silhouette_samples
 
silhouette_score(X, labels, metric='euclidean')
```

```python
# 군집별 실루엣 계수 계산
score_samples = silhouette_samples(X, data['kmeans_cluster'])
data['kmeans_silhouette_coeff'] = score_samples

data.groupby('kmeans_cluster')['kmeans_silhouette_coeff'].mean()
```

```python
# 군집별 관측치 개수 확인
data["kmeans_cluster"].value_counts()
```

# GMM
* GMM은 Parametric 모델로서 EM(Expectation Maximization) 알고리즘을 이용한 대표적인 군집화 모델
* 개별 데이터가 가우시안 분포에 속한다고 가정을 한 상태에서 특정 정규분포에 속할 확률을 추정하는 것이다. 
* n_components: 군집화 개수 

```python
from sklearn.mixture import GaussianMixture

# n_components로 미리 군집 개수 설정
gmm = GaussianMixture(n_components=6)
labels = gmm.fit_predict(X)

# GMM 후 클러스터링 레이블을 따로 설정
data['gmm_cluster'] = labels

data.head()
```

```python
# 전체 실루엣 계수 계산
from sklearn.metrics import silhouette_score, silhouette_samples
 
silhouette_score(X, labels, metric='euclidean')
```

```python
# 군집별 실루엣 계수 계산
score_samples = silhouette_samples(X, data['gmm_cluster'])
data['GMM_silhouette_coeff'] = score_samples

data.groupby('gmm_cluster')['GMM_silhouette_coeff'].mean()
```

```python
# 군집별 관측치 개수 확인
data["gmm_cluster"].value_counts()
```
