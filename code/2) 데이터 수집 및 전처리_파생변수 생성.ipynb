```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/응통 분석 공모전/데이터/"
```

```python
# 행정동별 카페 수, 편의점 수가 담긴 데이터셋 불러오기기
cafe_conv = pd.read_csv(path+"cafe_conv.csv")
cafe_conv = cafe_conv.rename(columns={"GU":"자치구", "DONG":"행정동"})
cafe_conv
```

```python
# 원본 데이터셋 불러오기기
data = pd.read_csv(path+'DATA_1.csv')
data
```

```python
del data["Unnamed: 0"]	
del data["Unnamed: 0.1"]
data = data.rename(columns={"GU":"자치구", "DONG":"행정동"})
data
```

```python
# 원본 데이터셋 내 카페 수, 편의점 수 col 제거하고 새 데이터셋과 concat 시키기기
del data["num_of_conv"]	
del data["num_of_cafe"]
data = pd.merge(data, cafe_conv, how='left', left_on=['자치구', '행정동'], right_on=['자치구', '행정동'])
data
```

```python
# 면목제3.8동에서 누락된 부분 채우고 면목3.8동 행 제거
data.loc[248, "자치구"] = "중랑구"
data.loc[248, "pop"] = 15440410.636
data.drop(424, axis=0, inplace=True)
data.drop(390, axis=0, inplace=True) # 항동 제거
data = data.reset_index(drop=True)
data
```

```python
# 서울시 구별 쓰레기 수거 현황 데이터 불러오기
collect = pd.read_csv(path+'쓰레기수거+현황_20221118163558.csv', skiprows=[0])
collect
```

```python
# 불필요한 행 제거 후 필요한 행만 선택하여 collect_concat에 저장 
collect.drop(0, axis=0, inplace=True)
collect_concat=collect[['자치구별(2)', '소계.1']]
collect_concat = collect_concat.rename(columns={"자치구별(2)":"자치구", "소계.1":"자치구별 배출량"})
collect_concat
```

```python
# 행정동별 데이터와 구별 전체 쓰레기 배출량 데이터 합치기 (자치구 기준)
data = pd.merge(data, collect_concat, how='left', left_on='자치구', right_on='자치구')
data
```

```python
# 행정동별 생활인구 열을 이용하여 자치구별 생활인구로 집계계
GU_pop = data.groupby('자치구')['pop'].sum().to_frame()
GU_pop = GU_pop.rename(columns={"pop":"자치구별 생활인구"})
GU_pop
```

```python
# 자치구별 생활인구 열 추가가
data = pd.merge(data, GU_pop, how='left', left_on='자치구', right_on='자치구')
data
```

```python
data = data.rename(columns={"pop":"행정동별 생활인구"})
data
```

```python
# 구별 쓰레기 배출량을 행정동별 생활인구로 곱한 뒤 자치구별 생활인구로 나누어 새로운 파생변수수 생성
data["행정동별 배출량"] = (data["자치구별 배출량"]*data["행정동별 생활인구"]) / data["자치구별 생활인구"]
data
```

```python
# 군집분석에서 불필요한 열 제거
data.drop(["자치구별 배출량",	"자치구별 생활인구"], axis=1, inplace=True)
data
```

```python
# 데이터셋 저장
data.to_csv(path+"data.csv", index=False)
```
